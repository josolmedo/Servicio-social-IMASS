```python
import matplotlib.pyplot as plt
import random

#G: Genuino
#I: No es genuino
numero_de_grabaciones=int(input("Ingrese el n√∫mero de grabaciones disponibles: "))

claves_genuino=[]
claves_no_genuino=[]

for grabaciones in range(numero_de_grabaciones):
  claves_genuino.append("G"+str(grabaciones+1))
  claves_no_genuino.append("I"+str(grabaciones+1))
print("")
print(claves_no_genuino)
print(claves_genuino)

probabilidad_similaridad_genuino=[]
probabilidad_tipicalidad_genuino=[]
probabilidad_similaridad_nogenuino=[]
probabilidad_tipicalidad_nogenuino=[]
probabilidad_genuino={}
probabilidad_no_genuino={}

for probabilidad in range(numero_de_grabaciones):
  probabilidad_similaridad_genuino.append(random.random())
  probabilidad_tipicalidad_genuino.append(random.random())
  probabilidad_similaridad_nogenuino.append(random.random())
  probabilidad_tipicalidad_nogenuino.append(random.random())

i=0
for grabaciones in range(numero_de_grabaciones):
  probabilidad_genuino[claves_genuino[i]]=probabilidad_similaridad_genuino[i]/probabilidad_tipicalidad_genuino[i]
  probabilidad_no_genuino[claves_no_genuino[i]]=probabilidad_similaridad_nogenuino[i]/probabilidad_tipicalidad_nogenuino[i]
  i=i+1

print("")
print(f"Probabilidad de similaridad en genuinos: {probabilidad_similaridad_genuino}")
print(f"Probabilidad de tipicalidad en genuinos: {probabilidad_tipicalidad_genuino}")
print(f"Probabilidad de similaridad en no genuinos: {probabilidad_similaridad_nogenuino}")
print(f"Probabilidad de tipicalidad en no genuinos: {probabilidad_tipicalidad_nogenuino}")
print("")
print("Likeihood ratio (score): ")
print(probabilidad_genuino)
print(probabilidad_no_genuino)

# Crear listas de probabilidades y etiquetas para el histograma
score_genuino = list(probabilidad_genuino.values())
score_no_genuino = list(probabilidad_no_genuino.values())

# Crear histograma
plt.hist(score_genuino, bins=30, color='blue', alpha=0.7, label='Genuino')
plt.hist(score_no_genuino, bins=30, color='red', alpha=0.7, label='No Genuino')

# Configurar el gr√°fico
plt.xlabel('Score')
plt.ylabel('Frecuencia')
plt.legend()
plt.title('Histograma grabaciones genuinas y no genuinas')
plt.show()
```
